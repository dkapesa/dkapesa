Feature Engineering

There are times when multiple features are provided in the same feature or we have to derive some features from the existing ones. I will also try to include some extra features in our dataset so, that we can derive some interesting insights from the data we have. Also if the features derived are meaningful then they become a deciding factor in increasing the modelâ€™s accuracy significantly.

```python
parts = df["date"].str.split("-", n = 3, expand = True)
df["year"]= parts[0].astype('int')
df["month"]= parts[1].astype('int')
df["day"]= parts[2].astype('int')
df.head()
```

```python
from datetime import datetime
import calendar

def weekend_or_weekday(year,month,day):

    d = datetime(year,month,day)
    if d.weekday()>4:
        return 1
    else:
        return 0

df['weekend'] = df.apply(lambda x:weekend_or_weekday(x['year'], x['month'], x['day']), axis=1)
df.head()
```

It would be nice to create a column which can indicate whether there was any holiday on a particular day or not.

```python
from datetime import date
import holidays

def is_holiday(x):

  india_holidays = holidays.country_holidays('IN')

  if india_holidays.get(x):
    return 1
  else:
    return 0

df['holdays'] = df['date'].apply(is_holiday)
df.head()
```

I will add some cyclical features

```python
df['m1'] = np.sin(df['month'] * (2 * np.pi / 12))
df['m2'] = np.cos(df['month'] * (2 * np.pi / 12))
df.head()
```

Adding a column whose value indicates which day of the week it is

```python
from datetime import datetime

def which_day(year, month, day):

    d = datetime(year,month,day)
    return d.weekday()

df['weekday'] = df.apply(lambda x: which_day(x['year'],
                                                    x['month'],
                                                    x['day']),
                                axis=1)
df.head()
```

Removing the columns which are not useful

```python
df.drop('date', axis=1, inplace=True)
```

Model Training

I will separate the features and target variables and split them into training and the testing data by using which we will select the model which is performing best on the validation data.

```python
features = df.drop(['sales', 'year'], axis=1)
target = df['sales'].values


X_train, X_val, Y_train, Y_val = train_test_split(features, target,
                                                  test_size = 0.05,
                                                  random_state=22)
X_train.shape, X_val.shape
```

Normalising the data before feeding it into machine learning models helps us to achieve stable and fast training.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

The data has been split into training and validation. Also the normalisation of the data has been done.

Now I will train some state-of-the-art machine learning models and select the best out of them using the validation dataset.

```python
models = [LinearRegression(), XGBRegressor(), Lasso(), Ridge()]

for i in range(4):
    models[i].fit(X_train, Y_train)

    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(Y_train, train_preds))

    val_preds = models[i].predict(X_val)
    print('Validation Error : ', mae(Y_val, val_preds))
    print()
```
